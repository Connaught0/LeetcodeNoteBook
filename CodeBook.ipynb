# 绘制点云

```python
s_pkl_file = '/home/Experient/SCARF/exps/snapshot/male-4-casual/male-4-casual/smpl_template.pkl'
S_dict = read_pickle(s_pkl_file)
x_dict = read_pickle('/home/baseline/gaussian-splatting/data/people_snapshot/male-4-casual/pixie/male-4-casual_f000000_param.pkl')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 创建点云数据
v = S_dict['verts']
vk = np.dot(v,x_dict['global_pose'].squeeze(0))

print(v.shape)
print(vk.shape)
 # 创建3D散点图
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

def render(v, paper, color):    
    # 提取X、Y和Z坐标
    x = v[:, 0]
    z = v[:, 1]  # 将Y坐标用于Z轴
    y = v[:, 2]  # 将Z坐标用于Y轴
    paper.scatter(x, y, z, c=color, marker='o')
# 绘制点云

render(v, ax, 'r')
render(vk, ax, 'b')

# 设置图形属性
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

# 将Y轴反转，使其朝上
ax.invert_yaxis()

# 显示图形
plt.show()
```
